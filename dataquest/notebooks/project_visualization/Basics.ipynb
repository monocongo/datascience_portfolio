### Do students in more popular majors make more money?
Let's plot sample size (popular majors have higher sample sizes) vs. median earnings to see.

```python
ax1 = recent_grads.plot(x='Sample_size',
                        y='Median', 
                        kind='scatter', 
                        title='Median vs. Sample_size', 
                        figsize=(5,10))
```

Hard to tell, probably not.

```python
ax2 = recent_grads.plot(x='Sample_size',
                        y='Unemployment_rate', 
                        kind='scatter', 
                        title='Unemployment_rate vs. Sample_size', 
                        figsize=(5,10))
ax3 = recent_grads.plot(x='Full_time',
                        y='Median', 
                        kind='scatter', 
                        title='Median vs. Full_time', 
                        figsize=(5,10))
ax4 = recent_grads.plot(x='ShareWomen',
                        y='Unemployment_rate', 
                        kind='scatter', 
                        title='Unemployment_rate vs. ShareWomen', 
                        figsize=(5,10))
ax5 = recent_grads.plot(x='Men',
                        y='Median', 
                        kind='scatter', 
                        title='Median vs. Men', 
                        figsize=(5,10))
```

### Do students that majored in subjects that were majority female make more money?
Let's plot vs. median earnings and see which side of the percentage the median earnings appear to be greater.

```python
ax7 = recent_grads.plot(x='ShareWomen',
                        y='Median', 
                        kind='scatter', 
                        title='Median vs. Percentage of Women', 
                        figsize=(5,10))
```

It appears that majors with a female majority earn less.

### Is there any link between the number of full-time employees and median salary?
Let's plot Full_time_year_round vs median earnings, to see if there's a positive correlation.

```python
ax8 = recent_grads.plot(x='Full_time_year_round',
                        y='Median', 
                        kind='scatter', 
                        title='Median vs. Full-time Employees', 
                        figsize=(5,10))
```

Again, hard to tell.

```python
ax = recent_grads['Median'].hist(bins=25, range=(0,recent_grads['Median'].max()))
ax.set_xlabel("Median Salary")
```

```python
def recent_grads_histogram(column_name):
    range_max = recent_grads[column_name].max()
    ax = recent_grads[column_name].hist(bins=25, 
                                        range=(0, range_max))
    ax.set_xlabel(column_name)
```

```python
recent_grads_histogram("Employed")
```

```python
recent_grads_histogram("Full_time")
```

```python
recent_grads_histogram("ShareWomen")
```

```python
recent_grads_histogram("Unemployment_rate")
```

```python
recent_grads_histogram("Men")
```

```python
recent_grads_histogram("Women")
```

```python
# sort the DataFrame by percentage of women
recent_grads.sort_values(by="ShareWomen", inplace=True)
ax1 = recent_grads[:10].plot.bar(x='Major', y='ShareWomen', title="Majors with the lowest percentage of women")
ax2 = recent_grads[-10:].plot.bar(x='Major', y='ShareWomen', title="Majors with the highest percentage of women")
```

```python
major_category_gender_counts = recent_grads.groupby(by=["Major_category"])[["Men", "Women"]].sum()
major_category_gender_counts["category"] = major_category_gender_counts.index
ax = major_category_gender_counts.plot.bar(x='category', 
                                           y=['Women', 'Men'], 
                                           title="Counts of women/men for major categories")
```
